```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("fitness.xlsx")
df=pd.DataFrame(data)
print(df)
plt.scatter(df["Flights Climbed"],df["Step Count"])
plt.xlabel("Flights Climbed",fontsize=20)
plt.ylabel("Step Count",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("fitness.xlsx")
df=pd.DataFrame(data)
print(df)
plt.bar(df["Flights Climbed"],df["Step Count"])
plt.xlabel("Flights Climbed",fontsize=20)
plt.ylabel("Step Count",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("fitness.xlsx")
df=pd.DataFrame(data)
print(df)
grouped_by=df.groupby("Step Count")["Distance"].sum()
print(grouped_by)
plt.plot(grouped_by.index,grouped_by.values)
plt.xlabel("Time",fontsize=20)
plt.ylabel("Distance",fontsize=20)
plt.show()
```
